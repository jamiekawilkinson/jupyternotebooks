```python
import pandas as pd
import numpy as np
```

```python
#Load in the dataset

df = pd.read_csv('scores.csv')
print(df)
```

```python
df.info()
```

```python
#clean dataset
```

```python
#1. Missing values dropped and save dataset

df2 = df.dropna(subset = ['Start Time', 'End Time','Student Enrollment', 'Percent White', 'Percent Black', 'Percent Hispanic',
                                'Percent Asian', 'Percent Tested'])
df2 = df2.reset_index(drop=True)

# Display the first few rows of the dataset
print(df2.head())
```

```python
df2.info()
```

```python
#For SAT scores, missing values will be inserted with averages of each subject

math_avg = df2['Average Score (SAT Math)'].mean()
reading_avg = df2['Average Score (SAT Reading)'].mean()
writing_avg = df2['Average Score (SAT Writing)'].mean()

df2['Average Score (SAT Math)'] = df2['Average Score (SAT Math)'].fillna(math_avg)
df2['Average Score (SAT Reading)'] = df2['Average Score (SAT Reading)'].fillna(math_avg)
df2['Average Score (SAT Writing)'] = df2['Average Score (SAT Writing)'].fillna(math_avg)
```

```python
df2.isna().sum()
```

```python
df2
```

```python
#Change start and end times to timedate data types
# Convert object columns to time data type
df2['Start Time'] = pd.to_datetime(df2['Start Time'], format='%H:%M:%S').dt.time
df2['End Time'] = pd.to_datetime(df2['End Time'], format='%H:%M:%S').dt.time
```
